```python
import os
import pandas as pd
import joblib
from androguard.core.bytecodes.apk import APK
```

```python
BASE_DATASET = r"C:\Users\SHAIKH\OneDrive\Desktop\ML\ML\dataset"
FEATURES_PATH = os.path.join(BASE_DATASET, "extracted_permission.csv")
MODEL_PATH = r"C:\Users\SHAIKH\OneDrive\Desktop\ML\ML\models\apk_classifier.pkl"
```

```python
import pickle

permission_columns_path = r"C:\Users\SHAIKH\OneDrive\Desktop\ML\ML\dataset/permission_columns.pkl"
permission_columns = pickle.load(open(permission_columns_path, "rb"))
```

```python
apk_path = input("Enter APK path: ")

if not os.path.exists(apk_path):
    raise FileNotFoundError("‚ùå APK file not found!")

print("\nüì¶ APK Loaded:", apk_path)
```

```python
try:
    apk = APK(apk_path)
    apk_permissions = apk.get_permissions()
    apk_permissions = [p.split(".")[-1] for p in apk_permissions]  # clean names
except Exception as e:
    print("‚ö† Error extracting permissions:", e)
    apk_permissions = []

print("\nüîç Extracted Permissions:")
print(apk_permissions)
```

```python
# -------------------------------
feature_vector = []

for perm in permission_columns:
    feature_vector.append(1 if perm in apk_permissions else 0)

X = pd.DataFrame([feature_vector], columns=permission_columns)
```

```python
model = joblib.load(MODEL_PATH)
```

```python
prediction = model.predict(X)[0]
proba = model.predict_proba(X)[0]

label = "Benign (SAFE)" if prediction == 0 else "Malware (DANGEROUS)"

print("\n===============================")
print("     üî• APK ANALYSIS RESULT üî•")
print("===============================")
print("Prediction:", label)
print("Confidence:", round(proba[prediction], 4))
print("===============================")
```
